# Typical ML pipeline order:

1) Data Collection
2) Data Cleaning (removing duplicates, irrelevant columns like IDs, etc., removing rows containiing NULL values, etc.)
3) EDA (Exploratory Data Analysis)

##### 4) Train-Test Split --> Split data into train and test
5) Filling missing values
6) Handling Outliers --> Detect outliers using training data only
7) Encoding (for categorical columns)
8) Function transformation (for numerical columns which are highly skewed)
9) Scaling numerical features
##### 9) Apply the same rules to test data (without learning from it)
10) Model Training
11) Optimization / tuning(Hyper Parameter Tuning)
12) Evaluation
##### 12) Deployment
13) Monitoring

# Should we remove outlier after or before test train split ?
One should usually remove or handle outliers AFTER the train-test split, using only the training data to decide what counts as an outlier.
##### Because,
If we remove outliers before splitting, information from the test set leaks into preprocessing. 
This causes data leakage, making model performance look better than it really is. The test set should remain completely unseen until evaluation.

# Should we apply Scaling after or before test train split ?
Scaling should also be applied AFTER the train-test split — and one should fit the scaler only on the training data.
##### Because,

If we scale before splitting, the scaler learns statistics (mean, std, min, max) from the whole dataset.
That leaks information from the test set into training.

# Should we handle missing values or duplicated values before split?
Cleaning operations(removing duplicates, irrelevant columns like IDs, etc., removing rows containiing NULL values, etc.) → can be BEFORE split

Statistic-learning operations → AFTER split

# Should we apply Function Transformation after the split?
Function transformation (like log, square root, Box-Cox) should generally be applied AFTER the train-test split, using only the training data 

# Is it necessary to apply Function Transformation?
Data is highly skewed; Large outliers dominate values

So,

1) For linear models → often helpful
2) For distance-based models → often helpful
3) For tree-based models → usually unnecessary

# Must Be Done After Train-Test Split

These steps calculate values from data, so they should learn only from training data.

1) Missing value imputation (mean/median/mode)
2) Scaling
3) Outlier treatment
4) Feature transformation
5) Encoding (usually fit on train)
6) Feature selection

# CORRECT ML PIPELINE

1) Raw Data
2) EDA
3) Basic Cleaning
4) Train-Test Split
5) Fit preprocessing on train
6) Apply same transformation on test
7) Model Training

# FEATURE ENGINEERING
Feature engineering is the process of creating, transforming, selecting, or improving features (columns) in your dataset so that a machine learning model can learn patterns better.

Feature engineering is used to improve the dataset for machine learning.
