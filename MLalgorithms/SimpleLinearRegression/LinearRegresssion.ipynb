# Simple Linear Regression
Linear Regression is used when I/P(Independent) data has LINEAR relationship with O/P(Dependent) data

Simple Linear Regression is used when there is only 'ONE independent variable' on which the 'dependent variable' depends

##### Algorithm based on: 
y = mx + c

```python
# Aim of the Experiment
# To build a machine learning model that can 'predict salary' based on 'years of experience' using Linear Regression.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Salary_Data.csv")
df.head(2)
```

```python
# Analyzing the dataset
# Need to apply regression here, coz O/P: Salary is 'CONTINUOUS NUMERICAL' in nature
```

```python
sns.scatterplot(x="YearsExperience", y="Salary", data=df)
plt.show()
```

### We observe that the graph follows simple linearity, so we should apply 'Simple Linear Regression'

```python
# checking for missing values
df.isnull().sum()
```

```python
# checking duplicate rows
df.duplicated().sum()
```

```python
# So, our Dataset is totally neat and clean
```

```python
df.info()
```

```python
# Test-Train Split
x = df[["YearsExperience"]] # We used [[]], coz we should APPLY ML algos. in 2D dataset
# If multiple columns were passed then there was no need to apply double brackets [[]]

# 1D will work for dependent variable 'y'
y = df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100) # 'random_state' to be changed(decrease or increase from 0 to so on based on dataset shape) if the accuracy improves
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# * Applying ML model

```python
from sklearn.linear_model import LinearRegression
# sklearn --> module
# linear_model --> sub module
# LinearRegression --> class
```

```python
# LinearRegression --> class, so we need to make its object
lr = LinearRegression()
lr.fit(x_train, y_train) # Whatever data is given to fit, i.e. to train, the model will be trained on y = mx + c
```

#### Training model on y = mx + c, means finding best value for 'm' & 'c' which fits the model, to give the 'BEST FIT LINE'
#### 'BEST FIT LINE' --> the line which passes from the very mid of the data points, 
#### in order to get the better prediction
#### PREDICTION IS DONE BASED ON THAT REGRESSION LINE 

# Checking Model Accuracy

```python
lr.score(x_test, y_test) * 100 # finding in %, so multiplying with 100
```

```python
# Gives best accuracy --> about 97% when 'random_state=100'
```

```python
lr.predict([[1.1]])
```
