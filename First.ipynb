# Introduction

Machine Learning (ML) is a branch of Artificial Intelligence (AI) where computers learn from pre-existing data and improve their performance without being explicitly programmed step-by-step.

Instead of writing exact rules, you give the computer data, and it learns PATTERNS from it.

# Types of ML

# SUPERVISED LEARNING
It is further classified into:

1) Regression Learning --> used when O/P data is 'continuous'
2) Classification Learning --> used when O/P data is 'discrete', (0 / 1), ('Y' /'N'), ('T', 'F'), ("Spam", "Not Spam"), etc.

# 1) Regression Algorithms
#### Linear Algorithms --> preferred when there is Linear relationship btw I/P & O/P features
1) Linear Regression
2) Multi-Linear Regression
3) Lasso Regression
4) Ridge Regression
#### Non-Linear Algorithms --> preferred when there is NOT the Linear relationship btw I/P & O/P features
1) Polynomial Regression
2) Decision Tree Regression
3) Random Forest Regression
4) Support Vector Machine
5) K-Nearest Neighbours
