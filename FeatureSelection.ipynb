# FEATURE SELECTION

Feature Selection is:
1) selecting the columns which are imp for prediction, 
2) removing those columns which are not that imp, 
3) merging two columns as one (if needed), etc.

To do feature selection, one should have:
1) knowledge of that domain

# FEATURE SELECTION TECHNIQUES
1) Forward Elimination (using MLxtend)
2) Backward Elimination (using MLxtend)

##### We can apply any of the above techniques, with 'k_features' and check the 'k_score'
##### Whichever set of columns gives highest accuracy, consider those set of columns(if rest not imp) for further prediction

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv("diabetes.csv")
df.head(3)
```

```python
x = df.iloc[:, :-1] # Input (Independent variables)
y = df["Outcome"] # Output (Dependent variables)
```

```python
x.shape[1] # accessing no. of rows
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

### Forward Elimination

```python
# Applying Forward Elimination, so 'forward=True'
k_sc = 0
k_ftrs = 0
k_ftrs_names = []
for i in range(1, x.shape[1]+1):
    
    fs = SequentialFeatureSelector(lr, k_features=i, forward=True)
    fs.fit(x, y)

    if(fs.k_score_ > k_sc):
        k_sc = fs.k_score_
        k_ftrs = i
        k_ftrs_names = fs.k_feature_names_

print("Best Score:", k_sc)
print("Best Number of Features:", k_ftrs)
print("Selected Features:", k_ftrs_names)
```
